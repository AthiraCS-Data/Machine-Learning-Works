Digits dataset from sklearn

```python
import pandas as pd
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
plt.show()
```

```python
df=pd.DataFrame(digits.data)
df.head()
```

```python
df["target"]=digits.target  # adding a new column called "target" including the data available in target directory
```

```python
df[0:12]
```

Train and model prediction

```python
x=df.drop("target",axis='columns')
y=df.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=20)     # number of decision tree here is given as 20, it can vary depends on the problem
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predicted=model.predict(x_test)
y_predicted
```

```python
len(x_test)
```

```python
len(y_train)
```

Confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predicted)
cm                                        #sum of the diagonals of cm will be equal to length of x_test
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('Truth')
plt.show()
```

Another method for plotting cm if the former one is not working

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm=confusion_matrix(y_test,y_predicted)
disp=ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='viridis')

plt.show()
```
