Problem statement : to find the prob of survival of females (titanic survival prediction)

```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
df
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
inputs=df.drop('Survived',axis='columns')
target=df.Survived
```

```python
# TO MAKE SEX COLUMN SEPERATE 
```

```python
dummies=pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
inputs=pd.concat([inputs,dummies],axis='columns')
inputs.head()
```

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs                                           # becoz we need only the prob of survival of females
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
