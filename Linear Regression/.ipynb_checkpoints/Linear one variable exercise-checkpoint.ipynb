```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.scatter(df.year,df.income,c="green",marker="o")
```

```python
df1=df.drop("income",axis='columns')
df1.head()
```

```python
df2=df.income
df2.head()
```

```python
model=linear_model.LinearRegression()
model.fit(df1,df2)
```

```python
model.predict([[2020]])
```

y=mx+c is the equation for linear regression, so we are finding intercept and coefficient to substitute in the equation 

```python
model.coef_
```

```python
model.intercept_
```
